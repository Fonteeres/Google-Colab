```python
import pandas as pd 
import numpy as np
df_vino = pd.read_csv('/content/drive/MyDrive/Aprendizaje_Automatico_1/set_datos/winequality-white.csv', sep=';')
df_vino
```

```python
df_vino_x = df_vino.drop(["quality"], axis = 1)
df_vino_x
```

```python
df_vino_y = df_vino['quality']
df_vino_y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_vino_x, df_vino_y, test_size=0.2, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
Arbol = DecisionTreeClassifier(random_state=45)
```

```python
Arbol.fit(X_train,y_train)
```

```python
Prediccion = Arbol.predict(X_test)
df_Prediccion = pd.DataFrame({'Actual':y_test, 'Prediccion':Prediccion})
df_Prediccion.head(70)
```

```python
score = Arbol.score(X_test, y_test)
score
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
```

```python
KNN.fit(X_train,y_train)
KNN_Predict = KNN.predict(X_test)
df_KNN_Predict = pd.DataFrame({'Actual':y_test, 'Prediccion':KNN_Predict})
df_KNN_Predict.head(70)
```

```python
KNN_score = KNN.score(X_test, y_test)
KNN_score
```

```python
k_range = range(1,20)
df_Resultados = pd.DataFrame(columns=['Rango','Resultado'])
for k in k_range:
  KNN2 = KNeighborsClassifier(n_neighbors=k)
  KNN2.fit(X_train,y_train)
  score = KNN2.score(X_test,y_test)
  df_Resultados.append({'Rango': k}) 
  df_Resultados.append({'Resultado': score})
  
df_Resultados
```
